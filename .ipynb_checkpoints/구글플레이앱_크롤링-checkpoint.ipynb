```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# ChromeDriver ì˜µì…˜ ì„¤ì •
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1280,720")  # ğŸ‘‰ í™”ë©´ í¬ê¸°ë¥¼ 1280x720ìœ¼ë¡œ ì¡°ì •
options.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])
options.add_argument("--disable-blink-features=AutomationControlled")

# ChromeDriver ì´ˆê¸°í™”
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# êµ¬ê¸€ í”Œë ˆì´ ìŠ¤í† ì–´ í˜ì´ì§€ë¡œ ì´ë™
url = "https://play.google.com/store/apps/details?id=com.sinhwa.tanda.riding&hl=ko&showAllReviews=true"
driver.get(url)

# **í˜ì´ì§€ ë¡œë“œ ëŒ€ê¸°**
time.sleep(5)  # ì „ì²´ ë¡œë”© ëŒ€ê¸°

# **ìŠ¤í¬ë¡¤í•˜ë©´ì„œ íŠ¹ì • ë²„íŠ¼ ë“±ì¥ ì—¬ë¶€ í™•ì¸**
scroll_pause_time = 1  # â³ ìŠ¤í¬ë¡¤ ê°„ê²© ì¡°ì • (1ì´ˆ)
max_scrolls = 20  # ìµœëŒ€ 20ë²ˆ ìŠ¤í¬ë¡¤ (í•„ìš”í•˜ë©´ ì¡°ì •)
scroll_distance = 500  # ğŸ’¡ ì¡°ê¸ˆì”© ìŠ¤í¬ë¡¤ (500pxì”©)

# ë²„íŠ¼ì˜ CSS Selector ë° XPath
button_css_selector = "#yDmH0d > c-wiz.SSPGKf.Czez9d > div > div > div:nth-child(1) > div > div.wkMJlb.YWi3ub > div > div.qZmL0 > div:nth-child(1) > c-wiz:nth-child(5) > section > header > div > div:nth-child(2) > button > i"
button_xpath = '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div/div[2]/div/div[1]/div[1]/c-wiz[5]/section/header/div/div[2]/button/i'

button_clicked = False  # ë²„íŠ¼ í´ë¦­ ì—¬ë¶€ í™•ì¸ ë³€ìˆ˜

for _ in range(max_scrolls):
    # ğŸ”½ ì¡°ê¸ˆì”© ë¶€ë“œëŸ½ê²Œ ìŠ¤í¬ë¡¤ (500pxì”© ì´ë™)
    driver.execute_script(f"window.scrollBy(0, {scroll_distance});")
    time.sleep(scroll_pause_time)

    try:
        # ë²„íŠ¼ì´ ë‚˜íƒ€ë‚¬ëŠ”ì§€ í™•ì¸ (CSS Selector ìš°ì„ )
        button = driver.find_element(By.CSS_SELECTOR, button_css_selector)
        
        if button.is_displayed():
            print("ğŸ”˜ ì¶”ê°€ ë¦¬ë·° ë²„íŠ¼ ë°œê²¬! í´ë¦­í•©ë‹ˆë‹¤.")

            # í´ë¦­í•˜ê¸° ì „ì— ë§ˆìš°ìŠ¤ ì»¤ì„œë¥¼ ë²„íŠ¼ ìœ„ë¡œ ì´ë™
            actions = ActionChains(driver)
            actions.move_to_element(button).perform()
            time.sleep(1)  # ë²„íŠ¼ í™œì„±í™” ëŒ€ê¸°
            
            # ë²„íŠ¼ í´ë¦­ (ë”ë¸” í´ë¦­ìœ¼ë¡œ í´ë¦­ ë³´ì¥)
            button.click()
            time.sleep(3)  # í´ë¦­ í›„ ë¡œë”© ëŒ€ê¸°

            # ğŸ”´ ë²„íŠ¼ì„ í´ë¦­í–ˆìœ¼ë¯€ë¡œ ìŠ¤í¬ë¡¤ ì¤‘ì§€
            button_clicked = True
            break

    except:
        try:
            # ë§Œì•½ CSS Selectorë¡œ ëª» ì°¾ìœ¼ë©´ XPathë¡œ ì°¾ê¸°
            button = driver.find_element(By.XPATH, button_xpath)
            
            if button.is_displayed():
                print("ğŸ”˜ ì¶”ê°€ ë¦¬ë·° ë²„íŠ¼ ë°œê²¬! (XPath) í´ë¦­í•©ë‹ˆë‹¤.")

                # ë§ˆìš°ìŠ¤ ì´ë™ í›„ í´ë¦­
                actions = ActionChains(driver)
                actions.move_to_element(button).perform()
                time.sleep(1)
                
                button.click()
                time.sleep(3)  # í´ë¦­ í›„ ë¡œë”© ëŒ€ê¸°

                # ğŸ”´ ë²„íŠ¼ì„ í´ë¦­í–ˆìœ¼ë¯€ë¡œ ìŠ¤í¬ë¡¤ ì¤‘ì§€
                button_clicked = True
                break

        except:
            pass  # ë²„íŠ¼ì´ ì•„ì§ ì•ˆ ë‚˜íƒ€ë‚˜ë©´ ê³„ì† ìŠ¤í¬ë¡¤

# ğŸ›‘ ë²„íŠ¼ í´ë¦­ í›„ ìŠ¤í¬ë¡¤ ë©ˆì¶¤
if button_clicked:
    print("âœ… ë²„íŠ¼ í´ë¦­ ì™„ë£Œ, ìŠ¤í¬ë¡¤ ì¤‘ë‹¨!")
else:
    print("âš  ë²„íŠ¼ì„ ì°¾ì§€ ëª»í–ˆê±°ë‚˜ í´ë¦­ ì‹¤íŒ¨.")

# í˜„ì¬ URL ì¶œë ¥í•˜ì—¬ ì •ìƒì ìœ¼ë¡œ ì—´ë ¸ëŠ”ì§€ í™•ì¸
print(f"í˜„ì¬ í˜ì´ì§€ URL: {driver.current_url}")

# WebDriver ì¢…ë£Œ
#driver.quit()
```
