```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# ChromeDriver 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1280,720")  # 👉 화면 크기를 1280x720으로 조정
options.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])
options.add_argument("--disable-blink-features=AutomationControlled")

# ChromeDriver 초기화
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# 구글 플레이 스토어 페이지로 이동
url = "https://play.google.com/store/apps/details?id=com.sinhwa.tanda.riding&hl=ko&showAllReviews=true"
driver.get(url)

# **페이지 로드 대기**
time.sleep(5)  # 전체 로딩 대기

# **스크롤하면서 특정 버튼 등장 여부 확인**
scroll_pause_time = 1  # ⏳ 스크롤 간격 조정 (1초)
max_scrolls = 20  # 최대 20번 스크롤 (필요하면 조정)
scroll_distance = 500  # 💡 조금씩 스크롤 (500px씩)

# 버튼의 CSS Selector 및 XPath
button_css_selector = "#yDmH0d > c-wiz.SSPGKf.Czez9d > div > div > div:nth-child(1) > div > div.wkMJlb.YWi3ub > div > div.qZmL0 > div:nth-child(1) > c-wiz:nth-child(5) > section > header > div > div:nth-child(2) > button > i"
button_xpath = '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div/div[2]/div/div[1]/div[1]/c-wiz[5]/section/header/div/div[2]/button/i'

button_clicked = False  # 버튼 클릭 여부 확인 변수

for _ in range(max_scrolls):
    # 🔽 조금씩 부드럽게 스크롤 (500px씩 이동)
    driver.execute_script(f"window.scrollBy(0, {scroll_distance});")
    time.sleep(scroll_pause_time)

    try:
        # 버튼이 나타났는지 확인 (CSS Selector 우선)
        button = driver.find_element(By.CSS_SELECTOR, button_css_selector)
        
        if button.is_displayed():
            print("🔘 추가 리뷰 버튼 발견! 클릭합니다.")

            # 클릭하기 전에 마우스 커서를 버튼 위로 이동
            actions = ActionChains(driver)
            actions.move_to_element(button).perform()
            time.sleep(1)  # 버튼 활성화 대기
            
            # 버튼 클릭 (더블 클릭으로 클릭 보장)
            button.click()
            time.sleep(3)  # 클릭 후 로딩 대기

            # 🔴 버튼을 클릭했으므로 스크롤 중지
            button_clicked = True
            break

    except:
        try:
            # 만약 CSS Selector로 못 찾으면 XPath로 찾기
            button = driver.find_element(By.XPATH, button_xpath)
            
            if button.is_displayed():
                print("🔘 추가 리뷰 버튼 발견! (XPath) 클릭합니다.")

                # 마우스 이동 후 클릭
                actions = ActionChains(driver)
                actions.move_to_element(button).perform()
                time.sleep(1)
                
                button.click()
                time.sleep(3)  # 클릭 후 로딩 대기

                # 🔴 버튼을 클릭했으므로 스크롤 중지
                button_clicked = True
                break

        except:
            pass  # 버튼이 아직 안 나타나면 계속 스크롤

# 🛑 버튼 클릭 후 스크롤 멈춤
if button_clicked:
    print("✅ 버튼 클릭 완료, 스크롤 중단!")
else:
    print("⚠ 버튼을 찾지 못했거나 클릭 실패.")

# 현재 URL 출력하여 정상적으로 열렸는지 확인
print(f"현재 페이지 URL: {driver.current_url}")

# WebDriver 종료
#driver.quit()
```
