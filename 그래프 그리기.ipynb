```python
import pandas as pd


country_data = pd.read_csv('./data/모든 국가_지역, 대한민국, 러시아, 튀르키예, 브라질, 태국, 미국, 독일, 베트남, 인도.csv')
country_data
```

```python
country_data = country_data.drop('참고사항', axis=1)
```

```python
country_data
```

```python
country_data['날짜'] = pd.to_datetime(
    country_data['날짜'].str.replace("년", "-", regex=True).str.replace("월", "-", regex=True).str.replace("일", "-", regex=True).str.replace(r"\s+", "", regex=True)
)
```

```python
country_data['날짜'] = pd.to_datetime(country_data['날짜'], format='%Y-%m-%d')
```

```python
country_data = country_data.rename(columns={
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 모든 국가/지역": "모든 국가/지역",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 대한민국": "대한민국",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 러시아": "러시아",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 튀르키예": "튀르키예",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 브라질": "브라질",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 태국": "태국",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 미국": "미국",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 독일": "독일",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 베트남": "베트남",
    "설치한 사용자 수 (모든 사용자 수, 순 사용자, 구간별, 일별): 인도": "인도",
})
```

```python
country_data = country_data.reset_index()  # 날짜를 다시 컬럼으로 복원
```

```python
# 쉼표 제거 후 숫자로 변환
for col in country_data.columns:
    country_data[col] = country_data[col].astype(str).str.replace(',', '').str.strip()

# 숫자로 변환 (변환 가능한 것만 변환)
for col in country_data.columns[1:]:  # 날짜 제외
    country_data[col] = pd.to_numeric(country_data[col], errors='coerce')
country_data["날짜"] = pd.to_datetime(country_data["날짜"])

# 변환 확인
print(country_data.info())
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib.dates as mdates

# 한글 폰트 설정 (Windows)
plt.rcParams["font.family"] = "Malgun Gothic"

# 날짜 컬럼을 인덱스로 설정
country_data["날짜"] = pd.to_datetime(country_data["날짜"])
country_data.set_index("날짜", inplace=True)

# 숫자 데이터 형식 변환 (쉼표 제거 후 정수형 변환)
country_data.iloc[:, 1:] = country_data.iloc[:, 1:].replace(',', '', regex=True).astype(int)

# 설치 수가 많은 상위 5개 국가 선택
top_countries = country_data.iloc[:, 1:].sum().nlargest(5).index

# 그래프 그리기
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.rcParams["font.family"] = "Malgun Gothic"

plt.figure(figsize=(12, 6))
for column in top_countries:
    plt.plot(country_data.index, country_data[column], marker="o", label=column)

plt.title("설치 사용자 수 변화 (상위 5개 국가)")
plt.xlabel("날짜")
plt.ylabel("설치 사용자 수")
plt.legend(title="국가")
plt.grid()

plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=3))

plt.show()

```
